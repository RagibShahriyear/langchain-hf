```python
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant. Please response to the user queries"),
        ("user","Question:{question}")
    ]
)
```

## Using the `HuggingFacePipeline`

```python
llm = HuggingFacePipeline.from_model_id(
    model_id="microsoft/phi-2",
    device = 0,
    task="text-generation",
    pipeline_kwargs={
        "max_new_tokens": 500,
        "top_k": 50,
        "temperature": 0.1,
    },
)
```

## Using the `HuggingFaceEndpoint`

```python
from langchain_huggingface import HuggingFaceEndpoint

llm = HuggingFaceEndpoint(
    repo_id="microsoft/Phi-3-mini-128k-instruct",
    task="text-generation",
    max_new_tokens=1024,
    do_sample=False,
    timeout=2000,
)
```

```python
output_parser=StrOutputParser()
chain=prompt|llm|output_parser
```

```python
response = chain.invoke({"Please write a push-pull-leg split for a male trying to build muscles, and add diet plan."})
print(response)
```
